```python
import pandas as pd
import random
```

```python
# Gerando uma lista de nomes fictícios
nomes = [f'Nome_{i}' for i in range(1, 51)]
```

```python
# Gerando uma lista para saber se é fumante ou não, de forma aleatória
fumante = [random.choice(['Sim', 'Não']) for _ in range(50)]
```

```python
# Criando o DataFrame
dataset = pd.DataFrame({
    'Nome': nomes,
    'Fumante': fumante
})
```

```python
# Mostrando os 30 primeiros registros do Dataset
dataset.head(30)
```

```python
import matplotlib.pyplot as plt
```

```python

# Contando quantos são fumantes e quantos não são
contagem_fumantes = dataset['Fumante'].value_counts()
```

```python
# Criando o gráfico de pizza
plt.figure(figsize=(6, 6))
plt.pie(contagem_fumantes, labels=contagem_fumantes.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Fumantes')
plt.show()
```
